# TASK 2 : VISUALISING THE TARGET(REVENUE) DISTRIBUTION

Trying to solve a regression problem, where our target is revenue to enable us make predictions.We use matloplip and seaborn to make side by side plots were we have the targetted distribution and the log transformed distribution
 

```python
#create subplot usint matplotlib
fig, ax=plt.subplots(figsize=(16,6))
#define layout of each subplots using seaborn
plt.subplot(1,2,1)
sns.distplot(train['revenue'], kde=False);
plt.title('Distribution of Revenue');
plt.subplot(1,2,2);
sns.distplot(np.log1p(train['revenue']), kde=False);
plt.title('Distribution of log-transformed revenue')
```

```python
#using the log tranformed is often better because it produces a kess skewed distribution. we add a log transformed revenue 
#to the data set
train['log_revenue']=np.log1p(train['revenue'])
```

# TASK 3 : REALTIONSHIP BEWTWEEN FILM BUDGET AND REVENUE

```python
#create 2 scatter plots  using seaborn 
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.scatterplot(x=train['budget'],y=train['revenue']);
plt.title('Revenue Vs Budget');
plt.subplot(1,2,2);
sns.scatterplot(x=np.log1p(train['budget']), y=train['log_revenue']);
plt.title('log Budget vs log Revenue')
```

```python
train['log budget']=np.log1p(train['budget'])
test['log budget']=np.log1p(test['budget'])
```

# TASK 4: EFFECT OF OFFICIAL HOMEPAGE ON MOVIE REVENUE

```python
#We calculate the hompage count per movie using the pandas value-count function
train['homepage'].value_counts().head(10)
```

```python
#create binary features for movies with homepage and those without homepage
train['has_homepage']=0
train.loc[train['homepage'].isnull()==False,'has_homepage']=1
test['has_homepage']=0
test.loc[test['homepage'].isnull()==False,'has_homepage']=1
```

```python
#creat a catplot using seaborn since we have classified the movies into 2 groups i.s those with homepage and those 
#without homepage
sns.catplot(x='has_homepage',y='revenue', data=train)
plt.title('Revenue for movies with or without home page')
```

# TASK 5: HOW DISTRIBUTION OF LANGUAGES AFFECT REVENUE

```python
#We locate the top 10 occuring lamguages in the 'original_language' column
language_data=train.loc[train['original_language'].isin(train['original_language'].value_counts().head(10).index)]
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.boxplot(x='original_language', y='revenue',data=language_data)
plt.title('mean revenue per language')
plt.subplot(1,2,2)
sns.boxplot(x='original_language',y='log_revenue',data=language_data)
plt.title('mean log revenue per language')
```
